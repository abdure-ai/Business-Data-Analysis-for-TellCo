```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

# Step 1: Descriptive Statistics
print("Descriptive Statistics:")
print(data.describe())

# Step 2: Missing Values and Outliers Treatment
# Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())

# Replace missing values with the mean or appropriate method
data.fillna(data.mean(), inplace=True)  # Replace with mean

# Step 3: Variable Transformations - Segment users by decile based on total session duration
data['Total Duration (ms)'] = data['Dur. (ms)']  # Ensure total duration is properly calculated
data['Decile Class'] = pd.qcut(data['Total Duration (ms)'], 10, labels=False)

# Calculate total data per decile class
data['Total Data'] = data['Total DL (Bytes)'] + data['Total UL (Bytes)']
decile_data = data.groupby('Decile Class').agg(
    total_data_per_decile=('Total Data', 'sum')
).reset_index()

# Visualize total data per decile class
plt.figure(figsize=(10, 6))
sns.barplot(x='Decile Class', y='total_data_per_decile', data=decile_data, palette='viridis')
plt.title('Total Data Usage per Decile Class')
plt.xlabel('Decile Class')
plt.ylabel('Total Data (Bytes)')
plt.show()

# Step 4: Univariate Analysis (e.g., Total DL and UL Data)
# Plot histogram and boxplot for 'Total DL (Bytes)'
plt.figure(figsize=(10, 6))
sns.histplot(data['Total DL (Bytes)'], kde=True, color='skyblue')
plt.title('Histogram of Total Download Data (Bytes)')
plt.xlabel('Total Download Data (Bytes)')
plt.ylabel('Frequency')
plt.show()

# Boxplot for Total DL (Bytes)
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['Total DL (Bytes)'], color='lightgreen')
plt.title('Boxplot of Total Download Data (Bytes)')
plt.xlabel('Total Download Data (Bytes)')
plt.show()

# Step 5: Bivariate Analysis (e.g., Social Media Data and Total Data)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['Social Media DL (Bytes)'], y=data['Total Data'], color='purple')
plt.title('Scatter Plot: Social Media DL (Bytes) vs Total Data')
plt.xlabel('Social Media Download Data (Bytes)')
plt.ylabel('Total Data (Bytes)')
plt.show()

# Step 6: Correlation Analysis
correlation_matrix = data[[
    'Social Media DL (Bytes)', 'Google DL (Bytes)', 'Email DL (Bytes)', 
    'Youtube DL (Bytes)', 'Netflix DL (Bytes)', 'Gaming DL (Bytes)', 
    'Other DL (Bytes)', 'Total DL (Bytes)', 'Total UL (Bytes)'
]].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# Step 7: Dimensionality Reduction using PCA
# Select the relevant columns for PCA
pca_data = data[[
    'Social Media DL (Bytes)', 'Google DL (Bytes)', 'Email DL (Bytes)', 
    'Youtube DL (Bytes)', 'Netflix DL (Bytes)', 'Gaming DL (Bytes)', 
    'Other DL (Bytes)', 'Total DL (Bytes)', 'Total UL (Bytes)'
]]

# Perform PCA
pca = PCA(n_components=2)  # Reduce to 2 components for visualization
pca_components = pca.fit_transform(pca_data)

# Plot PCA results
plt.figure(figsize=(10, 6))
plt.scatter(pca_components[:, 0], pca_components[:, 1], alpha=0.7, color='orange')
plt.title('PCA of Data Usage')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

# PCA Explained Variance
print("\nExplained Variance Ratio by PCA components:")
print(pca.explained_variance_ratio_)
```
